```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def proportional_area(no_of_points=1000000, dimension=2):
    # first we generate no_of_points random points in the unit hypercube
    points = np.random.uniform(-1, 1, size=(no_of_points, dimension))
    
    # than we calculate the distance of each point from the origin
    distance_of_points = np.linalg.norm(points, axis=1)
    
    # than we count the number of points inside the unit circle
    points_in_circle = np.sum(distance_of_points < 1)
    
    # than we calculat the proportion of points inside the unit circle
    proportion_of_points = (points_in_circle / no_of_points) 
    return proportion_of_points

# print(proportional_area(n=2))

# print(proportional_area(n=3))

# print(proportional_area(n=4))
```

```python
proportioanal_area_list = []

for i in range(2, 15):
    proportioanal_area_list.append(proportional_area(dimension=i))

plt.plot(range(2, 15), proportioanal_area_list)
plt.show()

print(proportioanal_area_list)
```
